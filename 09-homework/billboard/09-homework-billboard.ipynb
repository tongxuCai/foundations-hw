```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.billboard.com/charts/hot-100"
response = requests.get(url)
doc = BeautifulSoup(response.text)
```

```python
rows=[]
songs=doc.find_all(class_='chart-list-item')

for song in songs:
    row={}
    row['rank']=song.find(class_='chart-list-item__rank').string.strip()
    row['artist']=song.find(class_='chart-list-item__artist').text.strip()
    row['song']=song.find(class_='chart-list-item__title-text').text.strip()
    rows.append(row)
df=pd.DataFrame(rows)
df
```

```python
df.to_csv("billboard.csv", index=False)
```
