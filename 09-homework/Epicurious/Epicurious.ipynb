```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response=requests.get('https://www.epicurious.com/search/cucumbers')
doc = BeautifulSoup(response.text)
```

```python
rows=[]
cucumbers=doc.find_all(class_='recipe-content-card')
for cucumber in cucumbers:
    row={}
    row['Tag']=cucumber.find(class_='tag').text
    row['Title']=cucumber.find(class_='hed').text

    try:
        row['Summary'] = cucumber.find(class_='dek').text
    except:
        pass
    
    row['Rating']=cucumber.find(class_='rating').text.split('/')[0]
    row['Would make again']=cucumber.find(class_='make-again-percentage').text
    row['Link']=cucumber.find('a')['href']
    rows.append(row)
df=pd.DataFrame(rows)
df
```

```python
#build a function 
def scrape_page(row):
    return('https://www.epicurious.com'+row['Link'])
```

```python
food_links=df.apply(scrape_page, axis=1)
```

```python
food_rows=[]
for url in food_links:
    #turn this into a dict 
    print('We are scraping', url)
    #scrape the url
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    ingredients=doc.find(class_='ingredients').text
    names=doc.find(itemprop='name')  
    page={}
    page['names']=names
    page['ingredients']=ingredients
    print(page)
    food_rows.append(page)
    
```

```python
scraped_df=pd.DataFrame(food_rows)
scraped_df
```

```python
df.merge(scraped_df,left_index=True, right_index=True)
```

```python
df.to_csv("Epicurious_cleaned.csv", index=False)
```
