```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response=requests.get('https://www.goodreads.com/list/show/6934.Science_Fiction_Books_by_Female_Authors')
doc = BeautifulSoup(response.text)
```

```python
rows=[]
reads=doc.find_all('tr')

for read in reads:
    row={}
    row['Rank']=read.find(class_='number').string
    row['Title']=read.find('span').text.split('(')[0].strip()
    row['Author']=read.find(class_='authorName__container').text.strip()
    row['Score']=read.find(class_='smallText uitext').find('a').string
    row['Votes']=read.find(class_='smallText uitext').find_all('a')[1].text.split(' ')[0].strip()
    row['Rating']=read.find(class_='minirating').text.split(' ')[1]
    rows.append(row)
df=pd.DataFrame(rows)
df
```

```python
df['score']=df.score.str.replace('score:','')
```
