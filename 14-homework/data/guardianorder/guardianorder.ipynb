```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.rcParams['pdf.fonttype'] = 42

%matplotlib inline
```

```python
df=pd.read_csv('order-of-australia.csv')
df['percent_female']=100-df['percent_male']
df.head()
```

```python
df.sort_values(by='field').plot(title='Proportion of Order of Australia awards by category and gender, 2018',
                                x='field',y=['percent_male','percent_female'],
                                kind='barh',figsize=(20,20))
```

```python
df.sort_values(by='field').plot(title='Proportion of Order of Australia awards by category and gender, 2018',
                                x='field',y=['percent_male','percent_female'],color=['dodgerblue', 'slategray'],
                                kind='barh',figsize=(20,20),grid=True)
```
